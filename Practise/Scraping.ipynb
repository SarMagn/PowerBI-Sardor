```python
import scrapy
from scrapy_playwright.page import PageMethod

class PlaywrightSpider(scrapy.Spider):
    name = "playwright_spider"
    start_urls = ["https://m.timesjobs.com/mobile/jobs-search-result.html?txtKeywords=python"]

    custom_settings = {
        "PLAYWRIGHT_BROWSER_TYPE": "chromium",  # Brauzer turi
        "PLAYWRIGHT_LAUNCH_OPTIONS": {"headless": True},  # Korinmas rejim
    }

    async def parse(self, response):
        page = response.meta["playwright_page"]
        await page.wait_for_load_state("networkidle")  # Sahifaning toliq yuklanishini kutish

        jobs = response.css(".srp-comp-logo span.logo-comp-name::text").getall()

        print(jobs)
```
